**a) Đọc	bộ	dữ	liệu	train,	test	của	MNIST	từ	Keras**

```python
from keras.datasets import mnist

# Đọc bộ dữ liệu train và test từ Keras
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

**b) In	ra	kích	thước	của	tập	train,	test:**

```python
# In ra kích thước của bộ dữ liệu
print("Kích thước của bộ dữ liệu train:")
print("Train images:", train_images.shape)
print("Train labels:", train_labels.shape)
print("\nKích thước của bộ dữ liệu test:")
print("Test images:", test_images.shape)
print("Test labels:", test_labels.shape)
```

**c) Chuẩn	hóa	các	giá	trị pixel	của	ảnh	về	đoạn	[0,1]**

```python
# Chuẩn hóa giá trị pixel của ảnh về đoạn [0,1]
train_images_normalized = train_images / 255.0
test_images_normalized = test_images / 255.0
```

**Chia	tập	train ban	đầu	thành	tập	train	và	tập	validaion theo	tỷ	lệ	8:2**

```python
from sklearn.model_selection import train_test_split

# Chia tập train thành train và validation
train_images, validation_images, train_labels, validation_labels = train_test_split(train_images_normalized, train_labels, test_size=0.2, random_state=42)

# In ra shape của các tập train và validation
print("Shape của tập train_images:", train_images.shape)
print("Shape của tập train_labels:", train_labels.shape)
print("Shape của tập validation_images:", validation_images.shape)
print("Shape của tập validation_labels:", validation_labels.shape)
```

**Xây	dựng	Neural	Network	gồm các layers:**

```python
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout

# Khởi tạo mô hình Sequential
model = Sequential()

# Flatten layer: chuyển đổi dữ liệu từ định dạng 2D thành vector 1D
model.add(Flatten(input_shape=(28, 28)))

# Dense layer: 128 neurons, hàm kích hoạt Relu
model.add(Dense(128, activation='relu'))

# Dropout layer: ngẫu nhiên tắt một số neuron để giảm overfitting
model.add(Dropout(0.2))

# Output layer: số lượng neurons là số class cần phân lớp, hàm kích hoạt softmax
model.add(Dense(10, activation='softmax'))

# Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Hiển thị cấu trúc của mô hình
model.summary()
```

**Biên dịch mô hình sử dụng optimizer: adam; loss: sparse_categorical_crossentropy;	metric: accuracy**

```python
# Biên dịch mô hình
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

**Huấn	 luyện	 mô	 hình	 dựa	 trên	 tập	 train,	 đánh	 giá	 với	 tập	 validation;**
**epoch: 5;	batch_size:	32.**

```python
# Huấn luyện mô hình
history = model.fit(train_images, train_labels, epochs=5, batch_size=32, validation_data=(validation_images, validation_labels))
```

**Vẽ	biểu	đồ	thể	hiện	accuracy,	loss	sau	mỗi	epoch.**

```python
import matplotlib.pyplot as plt

# Lấy dữ liệu accuracy và loss từ biến history
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

# Vẽ biểu đồ accuracy
plt.plot(range(1, len(accuracy) + 1), accuracy, label='Training Accuracy')
plt.plot(range(1, len(val_accuracy) + 1), val_accuracy, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Vẽ biểu đồ loss
plt.plot(range(1, len(loss) + 1), loss, label='Training Loss')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

**Dự	đoán	phân lớp	cho	tập	test**

```python
import numpy as np

# Dự đoán xác suất cho từng class cho tập test
predictions = model.predict(test_images_normalized)

# Chọn ra class có xác suất lớn nhất cho mỗi mẫu dữ liệu
predicted_labels = np.argmax(predictions, axis=1)

# In ra dự đoán cho một số mẫu
print("Dự đoán phân lớp cho tập test:")
print(predicted_labels)
```

```python
from sklearn.metrics import accuracy_score

# Tính accuracy
accuracy = accuracy_score(test_labels, predicted_labels)

# In ra accuracy
print("Accuracy của mô hình trên tập test:", accuracy)
```

```python
from sklearn.metrics import classification_report

# Tính Precision, Recall, và F1-score
report = classification_report(test_labels, predicted_labels)

# In ra kết quả
print("Precision, Recall, và F1-score của mô hình trên tập test:")
print(report)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Tính confusion matrix
conf_matrix = confusion_matrix(test_labels, predicted_labels)

# Vẽ heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.arange(10), yticklabels=np.arange(10))
plt.title('Confusion Matrix')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python
# Lấy dự đoán phân lớp cho 5 ảnh đầu tiên trong tập test
for i in range(5):
    # Lấy ảnh từ tập test
    image = np.expand_dims(test_images_normalized[i], axis=0)

    # Dự đoán phân lớp
    prediction = model.predict(image)
    predicted_label = np.argmax(prediction)

    # In kết quả
    print("Predicted label for image", i+1, ":", predicted_label)

    # Hiển thị ảnh
    plt.figure()
    plt.imshow(test_images[i], cmap='gray')
    plt.title("Predicted label: " + str(predicted_label))
    plt.axis('off')
    plt.show()
```

**In	ra	05	kết	quả	đầu	tiên	phân	lớp	sai	so	với	nhãn	thực	tế	trong	tập	test**

```python
# Lấy chỉ số của các dự đoán sai
incorrect_indices = np.where(predicted_labels != test_labels)[0]

# In ra 5 kết quả phân lớp sai đầu tiên
for i in range(5):
    # Lấy chỉ số của ảnh phân lớp sai
    index = incorrect_indices[i]

    # Lấy ảnh từ tập test
    image = test_images_normalized[index]

    # Dự đoán phân lớp
    prediction = predicted_labels[index]

    # In kết quả
    print("Predicted label for image", index, ":", prediction)

    # Hiển thị ảnh
    plt.figure()
    plt.imshow(test_images[index], cmap='gray')
    plt.title("Predicted label: " + str(prediction))
    plt.axis('off')
    plt.show()
```
